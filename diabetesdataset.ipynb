```python
import pandas as pd
```

```python
df=pd.read_csv('diabetes.csv')
```

```python
df.head()
```

```python
x=df.iloc[:,:-1]
```

```python
x
```

```python
y=df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_text,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
model=Sequential()
```

```python
model.add(Dense(units=10,activation='relu',input_dim=8))
```

```python
model.add(Dense(units=8,activation='relu'))
```

```python
model.add(Dense(units=1,activation='sigmoid'))
```

```python
model.get_weights()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics='accuracy')
```

```python
model.fit(x_train,y_train,epochs=500)
```

```python
model.get_weights()
```

```python
predictions=model.predict(x_text)
```

```python
predictions
```

```python
class_labels=[]
```

```python
for i in range(predictions.size):
  if(predictions[i]>0.51):
    class_labels.append(1)
  else:
    class_labels.append(0)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(class_labels,y_test))
```
